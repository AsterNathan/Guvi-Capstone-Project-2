```python
# FILLING THE NUMERICAL VALUES
df['start_station_id'].fillna(df['start_station_id'].mean(), inplace = True)
df['end_station_id'].fillna(df['end_station_id'].mean(), inplace = True)
df['end_lat'].fillna(df['end_lat'].mean(), inplace = True)
df['end_lng'].fillna(df['end_lng'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
#FILLING THE NULL VALUES
df['start_station_name'].fillna(df['start_station_name'].mode()[0], inplace = True)
df['end_station_name'].fillna(df['end_station_name'].mode()[0], inplace = True)
```

```python
df.isnull().sum()
```

```python
df.head(15)
```

```python
#ADDING A NEW COLUMN CALLED trip_duration
# calculated by subtracting variables started_at from ended_at. However, both of these variables are in object type.
#So the first step is to convert them to date and time format using the function to_datetime() from the Pandas library. 
#The second step is subtracting variables started_at from ended_at.

pd.options.mode.chained_assignment = None  # default='warn'

df["started_at"] = pd.to_datetime(df["started_at"])
df["ended_at"] = pd.to_datetime(df["ended_at"])

print("Typecast of variables changed successfully.") 
df.info()
```

```python
df["trip_duration"] = df["ended_at"] - df["started_at"]
df["trip_duration"] = pd.to_numeric(df["trip_duration"])/6e+10
```

```python
df["trip_duration"].describe(include = "all")
```

```python
# removing outliers by creating filters
outliers_trip_duration = (df["trip_duration"] < 1) | (df["trip_duration"] > 480)
outliers_trip_duration.value_counts()
```

```python
df.drop(df.index[outliers_trip_duration], inplace = True)

print("Inconsistent values deleted successfully.")
```

```python
#ADDING A NEW COLUMN CALLED trip_distance
#calculated by a combination of starting and ending stations' GPS coordinates. The first step is using the function np.sqrt() from the NumPy library.
#The second step is multiplying the degrees of separation of longitude and latitude by 111.139 to get the corresponding linear distances in meters.
df["trip_distance"] = np.sqrt(((df["end_lat"] - df["start_lat"])**2) + ((df["end_lng"] - df["start_lng"])**2))
df["trip_distance"] = df["trip_distance"] * 111.139
```

```python
#ADDING A NEW COLUMN CALLED start_hour
#calculated from the variable started_at using the dt.hour function from the Datetime library.
df["start_hour"] = df["started_at"].dt.hour
df["start_hour"].unique()
```

```python
#ADDING A NEW COLUMN start_day
#calculated using the variable started_at with a combination of dictionaries and functions;
# dt.dayofweek from the Datetime library, and lambda.
df["start_day_num"] = df["started_at"].dt.dayofweek
day_dict = {0:"Sunday", 1:"Monday", 2:"Tuesday", 3:"Wednesday", 4:"Thursday", 5:"Friday", 6:"Saturday"}
df["start_day_name"]= df["start_day_num"].apply(lambda y:day_dict[y])
```

```python
df["weekend"] = np.where(df["start_day_num"] > 5, True, False)
```

```python
csv_data = df.to_csv('cyclistic_dataset.csv', index=False)
print("Dataset exported as CSV file successfully.")
```
